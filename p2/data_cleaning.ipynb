```python
import pandas as pd
```

Analytical Results data set columns:

**Sample ID#** - PDP Sample ID (unique sample identifier) - combination of 2-letter Collection State code, 2-digit Year, 2-digit Month, 2-digit Day, 4-digit Collection Site code, 2-letter Commodity code, 3-character Analyzing Lab code, and optional 1-letter Source ID code.9

**Com** - Commodity Code (2-letters) for product collected. Example: AP=Apples

**Pest Code** - FDA/PDP Pesticide Code (3-characters) for compound tested. Example: 011=Captan

**Pest Name** - Pesticide/compound Name

**Tst** - Code for Test Classification. Example: O=Pyrethroids

**Concen** - Concentration of residue detected. Zero (0) for non-detect.

**LOD** - Limit of Detection for pesticide/commodity pair

**pp_** - Concentration/LOD unit of measure (M=parts-per-million, B=pp-billion, T=pp-trillion)

**Co1** - Code for primary Confirmation method (1- or 2-letters). Example: LU=LC-MS/MS-triple quad

**Co2** - Code for secondary Confirmation method (1- or 2-letters)

**Ann** - Code for Annotated Information (1- or 2-letters). Example: Q=Below Quantifiable Level

**Qua** - Code for Quantitative method (1- or 2-letters). Example: E=Estimate

**Mea** - Code for Mean Result finding (1- or 2-letters). Examples: ND=Non-Detect

**Ext** - Code for Extraction method (3-digits). Example: 805=MDA Modified QuEChERS Method

**Det** - Code for Determinative method (2-digits). Example: 35=GC/MS/MS - triple quadrupole

**Tol (ppm)** - Established EPA Tolerance Level for pesticide/commodity pair in parts-per-million (ppm). Examples: 1.0 = 1.0 ppm, NT=No Tolerance; AL=FDA Action Level

```python
pdp = pd.read_csv('USDA_PDP_AnalyticalResults.csv')
pdp.head()
```

```python
pdp['State'] = pdp['Sample ID'].str[0:2]
pdp['Year'] = pdp['Sample ID'].str[2:4]
pdp['Month'] = pdp['Sample ID'].str[4:6]
pdp['Day'] = pdp['Sample ID'].str[6:8]
pdp['Collection Site'] = pdp['Sample ID'].str[8:12]
pdp['Community'] = pdp['Sample ID'].str[12:14]
pdp['Analyzing Lab'] = pdp['Sample ID'].str[14:17]
pdp['Source ID'] = pdp['Sample ID'].str[17:18]
pdp.head()
```

```python
pdp = pdp[['Sample ID', 'State', 'Year', 'Concentration', 'Pesticide Code', 'Pesticide Name']]
pdp['Year'] = pdp['Year'].astype(int)
pdp['State'] = pdp['State'].astype('string')
pdp = pdp.rename(columns = {'Sample ID': 'sample_id', 
                            'State': 'state',
                            'Year': 'year',
                            'Concentration': 'concentration',
                            'Pesticide Code': 'pesticide_code',
                            'Pesticide Name': 'pesticide_name'})
pdp.head()
```

```python
pdp = pdp[~pdp['year'].isin([94, 95, 96, 21, 22, 23])].reset_index(drop=True)
pdp.year.unique().tolist()
```

```python
monarchs = pd.read_csv('JNorthMonarch_with_fips.csv')
monarchs['STATEABBR'] = monarchs['STATEABBR'].astype('string')
monarchs = monarchs.rename(columns={'STATEABBR': 'state',
                                    'STATEFP': 'state_fips',
                                    'COUNTYFP': 'county_fips'})
monarchs = monarchs[(monarchs['species'] != 'Milkweed Sighted') & (monarchs['species'] != 'Milkweed (FIRST sighted)')]
monarchs['year'] = monarchs['sighting_date'].str[0:4]
monarchs = monarchs[['id', 'year', 'state', 'number']]
monarchs['year'] = monarchs['year'].astype(str).str[2:].astype(int)
monarchs = monarchs[~monarchs['year'].isin([94, 95, 96, 21, 22, 23])].reset_index(drop=True)
monarchs.head()
```
